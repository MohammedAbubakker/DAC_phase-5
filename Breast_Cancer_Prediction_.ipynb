```python
#converting words into numerical values
from sklearn.preprocessing import LabelEncoder
Le=LabelEncoder()
ds["diagnosis"]=Le.fit_transform(ds["diagnosis"])
```

```python
ds.corr()
```

```python
#setting the target variable and feature variable
from sklearn.model_selection import train_test_split
x=ds.drop(["diagnosis"],axis=1)
y=ds["diagnosis"]
```

```python
#splitting the dataset into train test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
#importing the algorithm
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
#prediction of testing dataset
prediction_knn=knn.predict(x_test)
print("prediction for testset;{}".format(prediction_knn))
```

```python
#finding the accuracy of the dataset
from sklearn.metrics import accuracy_score
print("accuracy for trainset:{:.2f}".format(knn.score(x_train,y_train)*100))
print("accuracy for testset:{:.2f}".format(knn.score(x_test,y_test)*100))
print("accuracy of the model:{:.2f}".format(accuracy_score(y_test,prediction_knn)*100))
```

```python
#finding the difference between actual value and predicted value
from sklearn.metrics import classification_report,confusion_matrix
diff_knn=pd.DataFrame({"Actual value":y_test,"predicted value":prediction_knn})
diff_knn
```

```python
#Visualizing the dataset using confusion matrix
con_mat=confusion_matrix(y_test,prediction_knn)
con_mat
```

```python
#importing matplotlit for plotting
import matplotlib.pyplot as plt
sns.heatmap(con_mat,annot=True,fmt="d")
plt.title("confusion matix")
plt.xlabel("predicted value")
plt.ylabel("actual value")
print(classification_report(y_test,prediction_knn))

```
